```python
%matplotlib inline
import os
from PIL import Image 
from pathlib import Path
import torch
import torchvision
torch.manual_seed(0)
print('Using PyTorch version', torch.version)


predict_transform = torchvision.transforms.Compose([
    torchvision.transforms.Resize(size=(224, 224)),
    torchvision.transforms.ToTensor(),
    torchvision.transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])
```

```python
model = torchvision.models.resnet50(pretrained=True)
model.fc = torch.nn.Linear(in_features=2048, out_features=3)
checkpoint = torch.load(Path('D:\\Project\\COVID Detection form Chest X-ray Images\\resnet18-model.h5'))
model.load_state_dict(checkpoint)

model.eval()
```

```python
image = Image.open("D:\\Project\\COVID Detection form Chest X-ray Images\\test\\COVID-5.png").convert('RGB')
image=predict_transform(image)
print(image.shape)
image=torch.unsqueeze(image, 0)
print(image.shape)
```

```python
output = model(image)
_, pred = torch.max(output, 1)
print(type(pred),pred.item())
```
